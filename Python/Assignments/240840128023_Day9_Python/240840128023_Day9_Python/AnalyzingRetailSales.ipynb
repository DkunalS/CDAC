```python
import pandas as pd
```

```python
df = pd.read_csv("retail_sales.csv")
df
```

```python
# Printing the top 10 product
df.head(10)
```

```python
# Date to datetime 
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
# Filtering the duplicates
filtered_duplicate = df.drop_duplicates() 
filtered_duplicate
```

```python
# Calculating the Quantifiers
Q1 = filtered_duplicate['Sales'].quantile(0.25)
Q3 = filtered_duplicate['Sales'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
# Determining the lower bound and upper bound
# Removing any duplicate entries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print(upper_bound)
median_sales = filtered_duplicate['Sales'].median()
filtered_duplicate['Sales'] = filtered_duplicate['Sales'].where((filtered_duplicate['Sales'] > lower_bound) & (filtered_duplicate['Sales'] < upper_bound), median_sales)
filtered_duplicate.head(20)
```

```python
# Calculating the total sales and printing the top 5 stores by sales 
total_sales = filtered_duplicate.groupby('Store')['Sales'].sum()
total_sales.head(5)
```

```python
# Calculationg the highest average sales
avg_sales = filtered_duplicate.groupby('Category')['Sales'].mean()
sorted_values = avg_sales.sort_values(ascending=False)
sorted_values.head(1)
```

```python
# Finding the best-selling product in terms of quantity sold
selling_prod = filtered_duplicate.groupby('Product')['Quantity'].sum()
best_selling_product = selling_prod.sort_values(ascending=False)
best_selling_product.head(1)
```

```python
# Using the IQR method to detect outliers in the Quantity column
outliers = filtered_duplicate[(filtered_duplicate['Quantity'] < lower_bound) | (filtered_duplicate['Quantity'] > upper_bound)]
outliers
```

```python
# Rank the products by total sales in descending order
prod_sales = filtered_duplicate.groupby('Product')['Sales'].sum()
ranking_product_bySales = prod_sales.rank()
ranking_product_bySales.sort_values(ascending=False)
```

```python
# Rank the stores by their average sales amount using the rank() method with the 'min' ranking method for ties
avg_sales_amount = filtered_duplicate.groupby('Store')['Sales'].mean().rank()
avg_sales_amount.rank(method='min')
```
