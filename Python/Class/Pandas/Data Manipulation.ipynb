# Data Manupulation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Cahrlie','lie', 'kob', 'Care'],
    'Age': [25,30,35,65,85,31],
    'City': ['New York', 'Los Angleles', 'Chicago','Nork', 'Loss', 'hicao'],
}
df = pd.DataFrame(data)
df
```

```python
df['Score'] = [55,66,88,22,44,33]
df
```

```python
df['Score'] = df['Score'] +2
df
```

```python
df = df.drop('City', axis = 1)
df
```

```python
sorted_age = df.sort_values(by = 'Age')
sorted_age
```

```python
sorted_index = df.sort_index(ascending = False)
sorted_index
```

```python
df = df.rename(columns = {'Score': 'Final Score'})
df
```
