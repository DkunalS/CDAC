```python
import pandas as pd
```

```python
#Raeding a Dataframe from series
df_student = pd.read_csv('student_data.csv')
df_student
```

```python
#Raeding a Dataframe from series
df_extra = pd.read_csv('additional_student_data.csv')
df_extra
```

```python
df_student = df_student.dropna(subset='Age')
df_student
```

```python
df_student = df_student.fillna(df_student['Age'].mean())
df_student
```

```python
filtered_data = df_student[(df_student['Score'] > 70) & (df_student['Age'] > 20)]
filtered_data
```

```python
filtered_data['Final_Score'] = filtered_data['Score'] + 10
filtered_data
```

```python
filtered_data = filtered_data.drop('Gender',axis=1)
filtered_data
```

```python
sorted_data = filtered_data.sort_values(by='Score',ascending=False)
sorted_data
```
