```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris Dataset from sns
df = sns.load_dataset('iris')
print(df.head()
```

```python
# scatter Plot : Petal length vs. Petal width

plt.figure(figsize=(8,6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df, palette='Set2', s=100)
plt.title('Scatter Plot: Petal Length vs. Petal Width')
plt.show()
```

```python
# Box Plot: Sepal Length by Species
plt.figure(figsize=(8,6))
sns.boxplot(x='species', y='sepal_length', data=df, hue='species', palette='Set3')
plt.title('Box Plot: Sepal Length by Species')
plt.show()
```

```python
# Violin Plot: Petal Length by Species
plt.figure(figsize=(8,6))
sns.violinplot(x='species', y='petal_length', data=df, palette='Pastel1')
plt.title('Violin Plot: Petal Length by Species')
plt.show()
```

```python
# Pair Plot : Relationship between all features
sns.pairplot(df, hue='species', palette='Set1')
plt.title('Pair Plot: Iris Dataset', y=1.02)
plt.show()
```

# Hands On

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
cols = [i for i in range(1,6)]
prod1 = np.random.randint(50,500,12)
prod2 = np.random.randint(50,500,12)
prod3 = np.random.randint(50,500,12)
prod4 = np.random.randint(50,500,12)
prod5 = np.random.randint(50,500,12)
```

```python
df = pd.DataFrame(data = {
    'Months': months,
    'prod1' : prod1,
    'prod2' : prod2,
    'prod3' : prod3,
    'prod4' : prod4,
    'prod5' : prod5
})
df
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x='Months', y='prod1', marker='o', data=df, palette='Pastel1', linestyle='--')
sns.lineplot(x='Months', y='prod2', marker='o', data=df, palette='Pastel2', linestyle='--')
sns.lineplot(x='Months', y='prod3', marker='o', data=df, palette='Pastel3', linestyle='--')
sns.lineplot(x='Months', y='prod4', marker='o', data=df, palette='Pastel4', linestyle='--')
sns.lineplot(x='Months', y='prod5', marker='o', data=df, palette='Pastel5', linestyle='--')

plt.title('Month: Product')
plt.show()
```

```python
sales = [a+b+c+d+e for a,b,c,d,e in zip(prod1,prod2, prod3, prod4, prod5)]
plt.figure(figsize=(20,10))
sns.barplot(months,sales, palette='bright',edgecolor='black', linewidth=2)
plt.title('Bar Chart Example')
plt.show()
```

```python
df = pd.DataFrame(df)
df_m = df.melt(id_vars=['Months'], var_name='Product', value_name='Sales')
sns.set(style="whitegrid")

plt.figure(figsize=(20, 10))
sns.barplot(x='Months', y='Sales', hue='Product', data=df_m, palette='colorblind')
plt.title('Monthly Sales by Product')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
sales = [a+b+c+d+e for a,b,c,d,e in zip(prod1,prod2, prod3, prod4, prod5)]
plt.figure(figsize=(15,13))
sns.violinplot(x=months, y=sales, data=df, palette='Pastel1')
plt.title('Violin Plot: Petal Length by Species')
plt.show()
```
