```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from matplotlib.dates import DateFormatter, AutoDateLocator
```

```python
# 1. Write a Pandas program to create a line plot of the historical stock prices of 
# Alphabet Inc. between two specific dates.

start_date = '2004-10-12'
end_date = '2005-06-07'
df = pd.read_csv('GooGL.csv')
df['Date'] = pd.to_datetime(df['Date'])
df_specified_date = df[df['Date'].between(start_date, end_date)]
df_specified_date.set_index('Date', inplace=True)
plt.figure(figsize=(10, 6))
plt.plot(df_specified_date['Close'], label='Alphabet Closing Prices', color='blue')
plt.title('Historical Stock Prices of Alphabet Inc. (GOOGLE)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 2.Write a Pandas program to create a line plot of the opening,
# closing stock prices of Alphabet Inc. between two specific dates.

plt.figure(figsize=(20,8))
plt.plot(df_specified_date['Open'],label="Opening Price",color = 'g',linestyle='--',marker='o')
plt.plot(df_specified_date['Close'],label='Closing Price',color='r',marker='p')
plt.title('Opening and Closing Price of Stock Price of Alphabet Inc.')
plt.xlabel('Dates')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 3. Write a Pandas program to create a bar plot of the trading volume of Alphabet 
# Inc. stock between two specific dates.

start_date1 = '2004-10-12'
end_date1 = '2004-11-12'

df_specified_date = df[df['Date'].between(start_date1, end_date1)]
df_specified_date.set_index('Date', inplace=True)

plt.bar(df_specified_date.index, df_specified_date['Volume'], label='Volume', color='blue', alpha=0.6)
plt.title('Historical Stock Prices of Alphabet Inc. (GOOGLE)')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 14. Write a Pandas program to create a plot of adjusted closing prices, thirty 
# days and forty days simple moving average of Alphabet Inc. between two specific 
# dates.

start_date2 = '2004-10-12'
end_date2 = '2007-03-10'

df_specified_date = df[df['Date'].between(start_date2, end_date2)]
df_specified_date.set_index('Date', inplace=True)

df_specified_date['30_SMA'] = df_specified_date['Adj Close'].rolling(window=30).mean()
df_specified_date['40_SMA'] = df_specified_date['Adj Close'].rolling(window=40).mean()

plt.figure(figsize=(12, 6))

plt.plot(df_specified_date.index, df_specified_date['Adj Close'], label='Adjusted Close', color='blue')
plt.plot(df_specified_date.index, df_specified_date['30_SMA'], label='30-Day SMA', color='orange')
plt.plot(df_specified_date.index, df_specified_date['40_SMA'], label='40-Day SMA', color='green')

plt.title('Adjusted Closing Prices and Moving Averages of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
```

```python
# 15. Write a Pandas program to create a plot of adjusted closing prices, 30 days 
# simple moving average and exponential moving average of Alphabet Inc. 
# between two specific dates.

df_specified_date['30_SMA'] = df_specified_date['Adj Close'].rolling(window=30).mean()
df_specified_date['30_EMA'] = df_specified_date['Adj Close'].ewm(span=30, adjust=False).mean()

plt.figure(figsize=(12, 6))

plt.plot(df_specified_date.index, df_specified_date['Adj Close'], label='Adjusted Close', color='blue')
plt.plot(df_specified_date.index, df_specified_date['30_SMA'], label='30-Day SMA', color='orange')
plt.plot(df_specified_date.index, df_specified_date['30_EMA'], label='30-Day EMA', color='green')

plt.title('Adjusted Closing Prices and Moving Averages of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.show()
```

```python
# 17. Write a Pandas program to create a plot to visualize daily percentage returns 
# of Alphabet Inc. stock price between two specific dates.


df_specified_date['Daily_Return'] = df_specified_date['Adj Close'].pct_change() * 100
df_specified_date.fillna({'Daily_Return': 0}, inplace=True)

plt.figure(figsize=(15,8))
plt.plot(df_specified_date.index, df_specified_date['Daily_Return'], label='Daily Return', color='blue', alpha=0.5)

plt.title('Adjusted Closing Prices and Daily Returns of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.xticks(rotation=90)

plt.legend()
plt.show()
```

```python
# 18. Write a Pandas program to plot the volatility over a period of time of Alphabet 
# Inc. stock price between two specific dates.


df_specified_date['Daily_Return'] = df_specified_date['Adj Close'].pct_change() * 100
df_specified_date.fillna({'Daily_Return': 0}, inplace=True)

# 21 trading days in a month
volatility = df_specified_date['Daily_Return'].rolling(window=21).std()

plt.figure(figsize=(15,8))

plt.figure(figsize=(15,8))
plt.plot(volatility, label='Volatility', color='blue', alpha=0.5)

plt.title('Adjusted Closing Prices and Daily Returns of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Volatility (%)')
plt.xticks(rotation=90)

plt.legend()
plt.show()
```

```python
# 19. Write a Pandas program to create a histogram to visualize daily return 
# distribution of Alphabet Inc. stock price between two specific dates

plt.figure(figsize=(15,8))
plt.hist(df_specified_date['Daily_Return'], bins=50, label='Daily Return', edgecolor='black', color='blue', alpha=0.5)

plt.title('Adjusted Closing Prices and Daily Returns of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Daily Return (%)')
plt.xticks(rotation=90)

plt.legend()
plt.show()
```
