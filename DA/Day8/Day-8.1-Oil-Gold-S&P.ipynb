```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SP500 oil gold bitcoin.csv')
df.head()
```

```python
#  Calculate correlation 
corr_bitcoin_gold = df[['BITCOIN','Gold']].corr()
corr_gold_SandP =  df[['Gold','S&P500']].corr()
corr_SandP_bitcoin = df[['S&P500','BITCOIN']].corr()
```

```python
#  Calculate Covariance 
# cov_bitcoin_gold = df[['BITCOIN','Gold']].cov()
# cov_gold_SandP =  df[['Gold','S&P500']].cov()
# cov_SandP_bitcoin = df[['S&P500','BITCOIN']].cov()
```

```python
fig ,axes = plt.subplots(1,3, figsize=(18,6))
# Bitoin and Gold data plot 
axes[0].scatter(df['BITCOIN'],df['Gold'],color='b',alpha=0.5)
axes[0].set_title(f'Bitcoin and Gold Correlation: {corr_bitcoin_gold.loc['BITCOIN','Gold']:.2f}')
axes[0].set_xlabel('BITCOIN')
axes[0].set_ylabel('Gold')
axes[0].grid(True)

# Gold and S&P500 data plot 
axes[1].scatter(df['Gold'],df['S&P500'],color='g',alpha=0.5)
axes[1].set_title(f'Gold and S&P500 Correlation: {corr_gold_SandP.loc['Gold','S&P500']:.2f}')
axes[1].set_xlabel('Gold')
axes[1].set_ylabel('S&P500')
axes[1].grid(True)

# S&P500 and Bitcoin data plot 
axes[2].scatter(df['S&P500'],df['BITCOIN'],color='r',alpha=0.5)
axes[2].set_title(f'S&P500 and Bitcoin Correlation: {corr_SandP_bitcoin.loc['S&P500','BITCOIN']:.2f}')
axes[2].set_xlabel('P500')
axes[2].set_ylabel('BITCOIN')
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
xmin, xmax = plt.xlim()
ymin, ymax = plt.ylim()
# Set the ts-ext position at 70% of the x-axis and 30% of the y-axis range
text_x = xmin + 0.7 * (xmax - xmin)
text_y = ymin + 0.3 * (ymax - ymin)
```

```python
plt.text(text_x, text_y, f'Correlation: {corr_bitcoin_gold.loc['BITCOIN','Gold']:.2f}', ha='center', va='center', fontsize=12)

# Horizontal alignment (ha): Controls how the text is positioned horizontally 
# in a relation to the x-coordinate, possible values: left, center , right position
# text vertically in the relation to the y-cordinates, possible values: top, center, bottom
plt.grid(True)
plt.show()
```
