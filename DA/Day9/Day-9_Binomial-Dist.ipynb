```python
import numpy as numpy 
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
df = pd.read_csv('tips.csv')
```

```python
# Binary outcome : success if ( tip / total bill ) > 0.15, otherwise failure
def classify_tip(row):
    total_bill = row['total_bill']
    tip = row["tip"]
    if (tip/total_bill) >0.15:
        return 1       # Success
    else:
        return 0       # Failure

# Apply the function to create the binary column
df['tips_binary'] = df.apply(classify_tip,axis = 1)

# Total number of trials ( total numbers observation)
n = len(df)

# Number of successes ( tips where tip/total_bill > 0.15 )
k = df['tips_binary'].sum()

# PROBABOILITY OF SUCCESS  (p)
p = k/n 

# define numbers of trials for the binomial distribution 
trials = 100 

# Genarate x valuse s(number of successes from 0 to 100)
x = np.arange(0, trials + 100)

# Calculate the PMF for each number of successes 
pmf_values = binom.pmf(x,trials,p)
cummulative_prob = 0
for i in x:
    cummulative_prob = binom.cdf(i,trials,p)
    print(f"probability for {i} success = {pmf_values[i]:.6f} ... Cumulative probability = {cummulative_prob:.6f} ")
```
