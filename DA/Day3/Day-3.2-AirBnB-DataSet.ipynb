```python
import numpy as np
import pandas as pd
```

```python
df_airbnb = pd.read_csv('AB_NYC_2019.csv')
df_airbnb
```

```python
# Make id column the index
df2 = df_airbnb.set_index('id')
df2.head(3)
df2.name[3647]        # Get the name for id = 3467  because we setted the index on id
```

```python
# Set pandas display options to show all colums which having float and int datatype
numeric_columns = df2.select_dtypes(include=['float64','int64'])
numeric_columns
```

```python
# Include/adding 'room_type' column in numeric_columns DataFrame
numeric_columns['room_type'] = df2['room_type']
numeric_columns
```

```python
# Here we are grouping all the roomtypes and related data
df3 = numeric_columns.groupby('room_type').mean()
df3
```

```python
# Here printing the max columns limit
pd.set_option('display.max_columns',None)
df3.head(3)
df3.index
```

```python
# Note that room_type has become an index - we can again convert it into a normal column
df3 = df3.reset_index()
pd.reset_option('display.max_columns')  # Also reset the column display to default
df3.head(3)
```
