```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
airline_df = pd.read_csv('airline_dec_2008_50k.csv')
distance = airline_df['Distance']
```

```python
# Calculte mean, median, skenwss , and kurtosis
mean_distanace = distance.mean()
median_distanace = distance.median()
skewness_distanace = distance.skew()
kurtosis_distanace = distance.kurt()   # Calculate kurtosis using 'kurt'
```

```python
# Visualize skewness with mean and median lines
plt.figure(figsize=(8,6))
sns.histplot(distance, kde=True, color='skyblue', bins=30)
plt.axvline(mean_distanace, color='red', linestyle='--', label=f'Mean:{mean_distanace:.2f}')
plt.axvline(median_distanace, color='black', linestyle='--', label=f'Median:{median_distanace:.2f}')
plt.title(f'Distribution of Distance\nSkewness: {skewness_distanace:.2f}, Kurtosis:{kurtosis_distanace:.2f}')
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```
