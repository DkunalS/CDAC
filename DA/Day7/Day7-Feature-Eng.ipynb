```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew
```

```python
# Generate random data with a larger sample size
sample_size = 10000
ratings = np.random.randint(1,6, sample_size)
```

```python
# Modify data for right skew
right_skew_data = np.concatenate((ratings, np.random.randint(1,2,sample_size//2)))
```

```python
# Modify data for right skew
left_skew_data = np.concatenate((ratings, np.random.randint(4,6,sample_size//2)))
```

```python
# Function to calculate mean, median, and skewnessabs
def calculate_stats(data):
    mean = np.mean(data)
    median = np.median(data)
    skewness = skew(data)
    return mean, median, skewness
```

```python
# Function to plot histogram with mean and median lines
def plot_histogram(ax, data, title):
    ax.hist(data, bins=np.arange(1,7) - 0.5, rwidth=0.8, alpha=0.7)
    ax.set_title(title)
    ax.set_xlabel('Rating')
    ax.set_ylabel('Frequency')
    ax.set_xticks(range(1,6))
    ax.grid(axis='y', alpha=0.5)

    # Calculate mean, median, and skewness
    mean, median, skewness = calculate_stats(data)
    ax.axvline(x=mean, color='red', linestyle='--', label=f'Mean:{mean:.2f}' )
    ax.axvline(x=median, color='blue', linestyle='--', label=f'Median: {median}')
    ax.legend()

    # Print mean, median, and skewness
    print(f'{title}')
    print(f'Mean: {mean:.2f}')
    print(f'Median: {median}')
    print(f'Skewness: {skewness: .2f}')
    print()  
```

```python
fig, axs = plt.subplots(1,3,figsize=(15,5))
plot_histogram(axs[0], ratings, 'No Skew')
plot_histogram(axs[1], right_skew_data, 'Right Skew Data')
plot_histogram(axs[2], left_skew_data, 'Left Skew')
plt.tight_layout()
plt.show()
```
