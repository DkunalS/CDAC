```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# **Scatter Plot**

```python
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
# Create scatter plot with matplotlib
plt.figure(figure=(8,6))
plt.scatter(x,y, color = 'blue', alpha = 0.6)
plt.title('Scatter plot using Matplotlib', fontsize=15)
plt.xlabel('X-axis',fontsize=12)
plt.ylabel('Y-axis',fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Create scatter using seaborn
sns.set(style="whitegrid")     # set styple with whitegrid background
plt.figure(figsize=(8,6))
sns.scatterplot(x=x, y=y)
plt.title('Scatter Plot using Seaborn', fontsize=15)
plt.show()
```

# **Line Plot**

```python
df = pd.read_csv("heart.csv")
ages = df.groupby('age').median().reset_index()
df.head()

# What [   ages = df.groupby('age').median().reset_index()   ] is doing?
#   => (1) Group on the age column
#      (2) Calculate median for each numeric column on the age group
#      (3) Reset index to reset to numeric index
```

```python
df.info()
```

```python
# Simple Pandas
ages.plot.line("age","chol");
plt.show()
```

```python
# Matplotlib
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html
fig,ax = plt.subplots()
ax.plot(ages["age"],ages["chol"], ls=":", lw=1.7)
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol");
plt.show()
```

```python
# Seaborn
sns.lineplot(x="age", y="chol", data=ages, linestyle=":", linewidth = 1.7)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

# **Count Plot**    
**# Also Konw as histogram**

```python
## Countplot: A simple plot, it merely shows the total count of rows per category
# A histogram across a categorical, insted of quantiative, variable
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
plt.figure(figsize=(10,4), dpi= 200)   # dpi =  Dots per Ince of space
sns.countplot(x = 'division', data = df)
plt.show()
```

```python
# Breakdown within another category with 'hue'
plt.figure(figure=(10,4),dpi=200)
sns.countplot(x = 'level of education', data = df, hue ='training level')
plt.show()
```

```python
# Using palette
plt.figure(figure=(10,4),dpi=200)
sns.countplot(x = 'level of education', data = df, hue ='training level',palette='cividis')
plt.show()
```

```python
plt.figure(figsize=(10,4),dpi = 200)
# Paired would be a good choice if there was a distinct jump from  0 and 1 to 2 and 3
sns.countplot(x = 'level of education', data = df, hue ='training level',palette='Paired')
plt.show()
```
